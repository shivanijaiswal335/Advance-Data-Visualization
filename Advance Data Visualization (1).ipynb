# Heat Map
   sns.heatmap()is used to create the heatmap.
   iris.corr()computes the correlation matrix of the Iris dataset using the        corr()function . This matrix contains pairwise corelation between the          numerical columns of dataset.
   annot = True displays the correlation values in each cell of the heatmap.
   cmap = 'cubehelix' sets the color map of the heatmap to 'cubehelix', which      is a colormap that varies smoothly between different colors.
   linewidths = 1 sets the width of the lines separating each cell in the          heatmap to 1
   linecolor = 'k' sets the color of the lines separating each cell to black .
   square = True makes each cell in the heatmap square-shaped. 
   mask = False does not mask any cells in the heatmap.
   vmin = -1 set the minimum and maximum values for the color scale,ensuring      that the heatmap ranges from -1 to 1 representing the correlation values.
   cbar_kws ={"orientation ":"vertical"} specifies the keyword arguments for      the colorbar. In the case , it sets the orientation of the colorbar to          vertical.
   cbar = True displays a colorbar next to the heatmap.
   

```python
fig = plt.gcf()
fig.set_size_inches(10,7)
fig = sns.heatmap(ir.corr(),annot=True,cmap='cubehelix',linewidths =1 , linecolor ='k',square=True, mask=False,vmin =-1, vmax=1)
```

```python
fig = plt.gcf()
fig.set_size_inches(12,8)
fig = sns.heatmap(ir.corr(),annot =True, cmap ="cubehelix",linewidths = 1,linecolor = "k",square =True, mask = False, vmin = -1,vmax = 1)
```

```python
# Swarm plot, like as stripplot
# A swarm plot, also known as a beeswarm plot, it is a type of categorical scatter plot
# That displays individual data points along an axis based on their categoridal variables.sns.set(style = "darkgrid")
sns.set(style = "darkgrid")
fig = plt.gcf()
fig.set_size_inches(10,7)
fig = sns.swarmplot(x = "Species",y="PetalLengthCm", data=ir)
```

```python
# To combine swarmplot and violinplot in same data 
# Define the figure size with plt than code for both
sns.set_style("darkgrid",{"axes.facecolor": ".9","grid.color":"0.2"})
fig = plt.gcf()
fig.set_size_inches(10,7)
ax = sns.violinplot(x="Species",y = "PetalLengthCm",data=ir,inner=None)
ax = sns.swarmplot(x="Species",y = "PetalLengthCm", data = ir , color="red",edgecolor = "black")
```
